<h3> 분포 범위가 비슷한 열끼리 합산한 파생변수 추가 </h3>

```python
# 열끼리의 합산을 반환하는 함수 
def get_sum_values(data_df, val_list):
  sums = []
  for i in range(data_df.shape[0]):
    vals = []
    for s in val_list:
      vals.append(data_df[str(s)].iloc[i])
    sums.append(sum(vals))
  return sums
```

<h3> 이상치 제거 </h3>

```python
# 데이터와 변수값을 할당받아 일정 범위를 벗어나면 해당 값을 0으로 대체
def get_outlier(df_raw=None, column=None):
  df_raw = pd.DataFrame(train_df[column])
  # 평균과 표준편차
  df_raw_mean = df_raw.mean(axis=0)
  df_raw_std = df_raw.std(axis=0)
  # mean - 3*std ~ mean + 3*std 범위 벗너나면 이상치로 제거
  df_raw_result = df_raw.copy()
  df_raw_result[df_raw_result > df_raw_mean + df_raw_std * 3] = 0
  df_raw_result[df_raw_result < df_raw_mean - df_raw_std * 3] = 0
  return df_raw_result
# 함수 사용해서 이상치 값 삭제
for i in range(1,len(train_df.columns)-4):
    rm_outlier = get_outlier(df_raw=train_df, column='X_'+str(i))
    train_df['X_'+str(i)]=rm_outlier
```

<h3> 처리 메소드 </h3>

```python
# 불필요한 열 제거 및 결측치 평균으로 대체 
def colDrop(pd):
    thr = len(pd)//2
    for i in pd.columns:
        if i in ["PRODUCT_CODE","LINE"]:
            continue
        if int(pd[i].isnull().sum()) > thr:
            pd.drop([i],axis = 1, inplace = True)
        else:
            pd[i].fillna(pd[i].mean(),inplace = True)
    return pd

# 불필요한 열 제거 및 결측치 평균으로 대체 
def TcolDrop(pd,pd2):
    col = pd.columns
    col = col.drop("Y_Quality")
    pd2 = pd2[col].copy()
    for i in pd2.columns:
        if i in ["PRODUCT_CODE","LINE"]:
            continue
        pd2[i].fillna(pd[i].mean(),inplace = True)
    return pd2
```
